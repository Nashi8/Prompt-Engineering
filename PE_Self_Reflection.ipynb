```python
!pip install --quiet openai

from google.colab import userdata
from openai import OpenAI

api_key = userdata.get("118A_OpenAI_API")

# Initialize OpenAI client
client = OpenAI(api_key=api_key)

print("OpenAI client initialized.")
```

```python
def call_llm(system_prompt, user_prompt, model="gpt-4.1-mini", temperature=0.3):
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user",  "content": user_prompt}
        ],
        temperature=temperature
    )
    return response.choices[0].message.content.strip()
```

```python
def summarize_text(text: str) -> str:
    system_prompt = "You are an excellent summarizer."
    user_prompt = f"""
Summarize the following text in 3–4 concise sentences.

TEXT:
{text}
"""
    return call_llm(system_prompt, user_prompt)


def critique_and_improve_summary(text: str, summary: str) -> str:
    system_prompt = "You are a careful writing critic and editor."
    user_prompt = f"""
You are given ORIGINAL TEXT and an INITIAL SUMMARY.

1. First, CRITIQUE the summary in bullet points. Focus on:
   - Accuracy (what is missing or incorrect?)
   - Clarity
   - Conciseness

2. Then, write an IMPROVED SUMMARY that:
   - Is accurate
   - Is clear
   - Is concise (3–4 sentences)

FORMAT YOUR RESPONSE LIKE THIS:

CRITIQUE:
- ...

IMPROVED SUMMARY:
<your improved summary here>

ORIGINAL TEXT:
{text}

INITIAL SUMMARY:
{summary}
"""
    return call_llm(system_prompt, user_prompt)
```

```python
print("Self-Reflection Summary Assistant")
print("Paste text to summarize, or type 'quit' to stop.")

while True:
    user_text = input("You (paste text, or 'quit'): ")
    if user_text.strip().lower() in {"quit", "exit"}:
        print("Assistant: Goodbye! ✨")
        break

    # Step 1: Initial summary
    initial_summary = summarize_text(user_text)
    print("\nAssistant - Initial Summary:\n")
    print(initial_summary)

    # Step 2: Critique + improved summary
    critique_output = critique_and_improve_summary(user_text, initial_summary)
    print("\nAssistant - Critique and Improved Summary:\n")
    print(critique_output)
    print("\n" + "-" * 60 + "\n")
```
